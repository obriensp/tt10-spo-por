```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd 
import seaborn as sns
 
df = pd.read_csv('temp.csv')
```

```python
sns.scatterplot(x='corner', y='delay', hue='temp', data=df, palette="icefire")
plt.show()
```

```python
fig, ax = plt.subplots()
sns.boxplot(x='corner', y='delay', hue='corner', order=['sf', 'ff', 'tt', 'ss', 'fs'], data=df.query('temp == 27'))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x*1000) + ' ms'))
ax.set_ylim(ymin=0)
plt.show()
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x='corner', y='delay', hue='temp', data=df.query('temp==27.0'), s=8, palette="icefire")
#ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x) + 'V'))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x*1000000) + 'us'))
ax.set_ylim(ymin=0)
ax.set_box_aspect(0.5)
fig.set_dpi(200)
plt.show()
```

```python
fig, ax = plt.subplots()
sns.lineplot(x='temp', y='delay', hue='corner', errorbar=lambda x: (x.min(), x.max()), data=df)
#sns.lineplot(x='temp', y='delay', errorbar=lambda x: (x.min(), x.max()), data=df)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x) + 'C'))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x*1000) + ' ms'))
ax.set_ylim(ymin=0)
ax.set_box_aspect(1)
ax.set_title('spo')
fig.set_dpi(150)
plt.show()
```
